```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
# 4. 多变量关联分析
# 使用多元线性回归模型分析
X = df[['包邮', '运费险', '新品', '公益宝贝', '全球购']]  # 独立变量
y = df['购买人数']  # 因变量
X = sm.add_constant(X)  # 添加常数项
model = sm.OLS(y, X).fit()
print(model.summary())

# 5. 统计检验
# 进行 T-检验比较包邮与非包邮的购买人数差异
t_stat, p_value = stats.ttest_ind(df[df['包邮'] == 1]['购买人数'],
                                  df[df['包邮'] == 0]['购买人数'])
print(f'T-statistic: {t_stat}, P-value: {p_value}')

# 6. 可视化
# 箱形图比较包邮与否的购买人数
plt.figure(figsize=(8, 5))
sns.boxplot(x='全球购', y='购买人数', data=df)
plt.title('%全球购购与否的购买人数箱形图')
plt.show()
```

```python
解释购买人数（因变量）如何受到包邮、运费险、新品、公益宝贝和全球购（自变量）的影响。下面是结果的一个简要解释：

R-squared (R²): R² 值为 0.007，这表明模型只解释了购买人数变异性的0.7%。这是一个非常低的值，意味着模型没有很好地捕捉到数据的变异性。

Adj. R-squared: 调整后的 R² 也很低，只有 0.004。这考虑了模型中变量的数量，是对 R² 的一种调整，以防止过多的自变量导致误导性的高 R² 值。

F-statistic: F统计量为2.309，其对应的概率（Prob (F-statistic)）为0.0421，这意味着模型中至少有一个预测变量对购买人数有统计学上的显著影响，因为这个值小于0.05。

P>|t|: 这一列显示的是每个变量的 t 检验的 p 值。通常，p 值小于0.05意味着该变量在统计上是显著的。

包邮: p 值为0.113，意味着包邮对购买人数的影响在统计上不显著。
运费险: p 值为0.378，也在统计上不显著。
新品: p 值为0.208，不显著。
公益宝贝: p 值为0.233，不显著。
全球购: p 值为0.006，这是唯一一个在统计上显著的变量，表明全球购有显著影响购买人数。
coef: 回归系数显示了当其他所有变量保持不变时，每个自变量与因变量之间的关系。例如，全球购的系数为-278.4847，这表明提供全球购的商品平均购买人数会比不提供全球购的商品少约278人。

Durbin-Watson: 这个统计量检测残差的自相关，其值接近于2表明残差没有自相关。这里的值为0.063，这可能表明数据存在正自相关问题，可能需要进一步诊断模型。

Omnibus/Prob(Omnibus): 检验残差是否正常分布的统计量，这里的p值接近0，表明残差不是正态分布的，可能违反了线性回归分析的正态性假设。

Jarque-Bera (JB)/Prob(JB): 另一个检验残差正态性的统计量，同样表明残差不是正态分布的。

Skew: 残差偏度，这里的值表明数据有很强的正偏斜。

Kurtosis: 残差峰度，这里的值非常高，表明数据的分布有厚尾。

Cond. No.: 条件数，用于检测多重共线性，这里的值不是很高，表明多重共线性可能不是主要问题。

综上所述，模型的解释能力较弱，只有全球购对购买人数有显著影响。由于残差的分布不是正态的，可能需要考虑转换响应变量或使用非参数方法
```

```python
from sklearn.ensemble import RandomForestRegressor

# 对响应变量进行对数变换
df['购买人数_log'] = np.log1p(df['购买人数'])

# 使用随机森林作为非参数回归模型
X = df[['包邮', '运费险', '新品', '公益宝贝', '全球购']]  # 独立变量
y = df['购买人数_log']  # 对数变换后的因变量

# 拟合随机森林模型
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# 输出特征的重要性
importance = rf.feature_importances_
print("Feature importances:", importance)

# 如果需要对未变换的购买人数进行预测，需要将结果变换回原始尺度
y_pred_log = rf.predict(X)
y_pred = np.expm1(y_pred_log)  # 使用 expm1 对数变换的逆变换
```

```python
特征重要性是在使用基于树的模型（例如随机森林）时衡量每个特征对模型预测能力影响的一种方式。在您提供的输出中，特征重要性的值表明了每个特征在模型中的相对重要性。这里是每个特征的重要性解释：

包邮 (0.17392638): 这个特征约占总重要性的17.39%，意味着它在所有特征中相对重要，但不是最重要的特征。

运费险 (0.30278008): 这是所有特征中最重要的，占总重要性的30.28%。这表明运费险在决定购买人数时起着较大作用。

新品 (0.12568694): 这个特征的重要性相对较低，约占12.57%，表明它对模型的贡献小于其他特征。

公益宝贝 (0.07382127): 这个特征的重要性最低，只占7.38%，可能对购买人数的预测贡献不大。

全球购 (0.32378533): 这个特征的重要性略高于运费险，约占32.38%，是影响购买人数的另一个主要因素。

根据这些特征重要性，您可以得出全球购和运费险是两个最重要的特征，对购买人数有显著影响。而公益宝贝的影响相对较小。
```

```python
from sklearn.model_selection import cross_val_score

# 使用交叉验证来评估模型性能
cv_scores = cross_val_score(rf, X, y, cv=5, scoring='neg_mean_squared_error')
cv_rmse = np.sqrt(-cv_scores)
print("CV RMSE scores:", cv_rmse)
print("CV RMSE mean:", cv_rmse.mean())

```

```python
from sklearn.metrics import mean_squared_error

# 计算训练集上的预测值
train_predictions = rf.predict(X)

# 计算 RMSE
train_mse = mean_squared_error(y, train_predictions)
train_rmse = np.sqrt(train_mse)
print("Train RMSE:", train_rmse)
```

```python
residuals = y - train_predictions

plt.figure(figsize=(10, 6))
plt.scatter(train_predictions, residuals)
plt.title('Residuals vs Predicted')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```
